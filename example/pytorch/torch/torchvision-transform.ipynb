```python
"""
    torchvision의 transform으로 이미지 정규화하기(평균, 표준편차를 계산하여 적용)

    일단 이해한 흐름
    > 이미지를 불러와서 Tensor형 데이터로 만든 다음에
    > 각 r, g, b 픽셀에 대한 최소, 최대, 평균, 표준편차 값을 구한다 (넘파이 개좋..)
    > 그 이후에 평균과 표준편차 값을 구한 걸 바탕으로 이미지 Normalize(평준화)를 한다
"""
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision
import torch.utils.data as torchdata
from torchvision import datasets, transforms, models
import matplotlib.pyplot as plt

batch_size = 4
classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')

transform = transforms.Compose([
    # transforms.Resize((224, 224)),
    transforms.ToTensor(), # tensor로 자료구조 변경
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)), # (평균), (표준편차) 인데 왜 예제는 다 0.5로 해놨을까나
    transforms.RandomResizedCrop(224), # 랜덤 이미지 크기 조정
    transforms.RandomHorizontalFlip(), # 랜덤 수직 뒤집기
    transforms.RandomRotation(30) # 랜덤 회전
])

train_set = datasets.CIFAR10(
    root="data",
    train=True,
    download=True,
    transform=transform
)
train_loader = torchdata.DataLoader(train_set, batch_size=batch_size, shuffle=True, num_workers=2)

test_set = datasets.CIFAR10(
    root="data",
    train=True,
    download=True,
    transform=transform
)
test_loader = torchdata.DataLoader(test_set, batch_size=batch_size, shuffle=True, num_workers=2)

def print_stats(dataset):
    imgs = np.array([img.numpy() for img, _ in dataset])
    print(f'shape: {imgs.shape}')

    # 각 이미지 RGB 픽셀의 최소값 구하기
    min_r = np.min(imgs, axis=(2, 3))[:, 0].min()
    min_g = np.min(imgs, axis=(2, 3))[:, 1].min()
    min_b = np.min(imgs, axis=(2, 3))[:, 2].min()

    # 각 이미지 RGB 픽셀의 최댓값 구하기
    max_r = np.max(imgs, axis=(2, 3))[:, 0].max()
    max_g = np.max(imgs, axis=(2, 3))[:, 1].max()
    max_b = np.max(imgs, axis=(2, 3))[:, 2].max()

    # 각 이미지 RGB 픽셀의 평균 구하기
    mean_r = np.mean(imgs, axis=(2, 3))[:, 0].mean()
    mean_g = np.mean(imgs, axis=(2, 3))[:, 1].mean()
    mean_b = np.mean(imgs, axis=(2, 3))[:, 2].mean()

    # 각 이미지 RGB 픽셀의 표준편차 구하기
    std_r = np.std(imgs, axis=(2, 3))[:, 0].std()
    std_g = np.std(imgs, axis=(2, 3))[:, 1].std()
    std_b = np.std(imgs, axis=(2, 3))[:, 2].std()

    print(f'RGB 최소값: {min_r, min_g, min_b}')
    print(f'RGB 최대값: {max_r, max_g, max_b}')
    print(f'RGB 평균값: {mean_r, mean_g, mean_b}')
    print(f'RGB 표준편차: {std_r, std_g, std_b}')

print_stats(train_set)
# model = models.inception_v3(pretrained=True)
# print(model)
# 학습된 모델 그대로 쓸 수 있긴한데 일단은 직접 만들고 다음에 비교해보자

class NeuralNetwork(nn.Module):
    def __init__(self):
        self.a = 1

print(torch.rand(2, 2))
```
